# Arithmetic Operators

```python
x = 158
y = 23

print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x // y)
print(x % y)
print(x ** 2)
```

# Comparison Operators

```python
print(23 < 34)
print(23 > 34)
print(23 == 34)
print(23 != 34)
print(23 <= 34)
print(23 >= 34)
```

# Logical operators

```python
# and
print(23 == 67 and 78 < 100) 
# or 
print(23 == 67 or 78 < 100) 
# not 
print( not 34 == 34)
```

# Assignment Opeators

```python
a = 10 
print(a)

# a = a + 10
a += 10
print(a)
a -= 10
print(a)
a *= 10
print(a)
a /= 10
print(a)
```

# Identity Operators

```python
#is 
a = 10
b = 20
print(a is b)
# is not
print(a is not b)
```

# Membership Operators

```python
web = "codeswithpankaj.com"
print('c' in web)
print('c' not in web)
```
